### **1.1 Latar Belakang**
<p> Matahari adalah bintang yang dinamis dan memiliki aktivitas magnetik yang berfluktuasi dalam periode tertentu. Salah satu indikator paling nyata dari aktivitas ini adalah munculnya bintik matahari (sunspots) di fotosfer. Bintik matahari merupakan area dengan suhu yang lebih rendah dibandingkan sekitarnya karena adanya konsentrasi medan magnet yang sangat kuat yang menghambat proses konveksi.</p> 

<p> Pengamatan bintik matahari telah dilakukan secara sistematis sejak abad ke-17. Data historis ini menunjukkan bahwa jumlah bintik matahari tidaklah acak, melainkan mengikuti pola periodik yang dikenal sebagai Siklus Schwabe. Memahami periode siklus ini sangat krusial bagi kehidupan di Bumi karena aktivitas matahari (seperti solar flare dan coronal mass ejection) dapat mempengaruhi cuaca antariksa, mengganggu komunikasi satelit, sistem GPS, hingga stabilitas jaringan listrik global.</p> 

<p> Dalam studi ini, kami menggunakan pendekatan Digital Signal Processing (DSP) dengan algoritma Fast Fourier Transform (FFT). Metode ini dipilih karena kemampuannya untuk mengubah data deret waktu (time-domain) yang penuh dengan gangguan (noise) menjadi domain frekuensi, sehingga kita dapat mengisolasi dan menentukan periode osilasi utama matahari secara matematis dan objektif.</p> 

<br>

### **1.2 Tujuan**
Tujuan dari tugas *case based* ini adalah:
1. Menganalisis data historis jumlah bintik matahari (*sunspot*) menggunakan metode *Fast Fourier Transform* (FFT).
2. Mengidentifikasi frekuensi dominan dari aktivitas matahari melalui transformasi data dari domain waktu ke domain frekuensi.
3. Menentukan periode utama siklus matahari berdasarkan hasil analisis spektrum frekuensi yang diperoleh.

<br>

### **1.3 Manfaat**
Manfaat dari tugas *case based* ini adalah:
* Menambah pemahaman tentang penggunaan *Fast Fourier Transform* (FFT) dalam analisis data deret waktu.
* Membantu memahami pola periodik aktivitas matahari melalui analisis domain frekuensi.
* Menjadi latihan penerapan konsep *Digital Signal Processing* (DSP) pada data nyata di bidang astrofisika.
