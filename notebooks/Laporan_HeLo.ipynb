### **1.1 Latar Belakang**
<p> Matahari adalah bintang yang dinamis dan memiliki aktivitas magnetik yang berfluktuasi dalam periode tertentu. Salah satu indikator paling nyata dari aktivitas ini adalah munculnya bintik matahari (sunspots) di fotosfer. Bintik matahari merupakan area dengan suhu yang lebih rendah dibandingkan sekitarnya karena adanya konsentrasi medan magnet yang sangat kuat yang menghambat proses konveksi.</p> 

<p> Pengamatan bintik matahari telah dilakukan secara sistematis sejak abad ke-17. Data historis ini menunjukkan bahwa jumlah bintik matahari tidaklah acak, melainkan mengikuti pola periodik yang dikenal sebagai Siklus Schwabe. Memahami periode siklus ini sangat krusial bagi kehidupan di Bumi karena aktivitas matahari (seperti solar flare dan coronal mass ejection) dapat mempengaruhi cuaca antariksa, mengganggu komunikasi satelit, sistem GPS, hingga stabilitas jaringan listrik global.</p> 

<p> Dalam studi ini, kami menggunakan pendekatan Digital Signal Processing (DSP) dengan algoritma Fast Fourier Transform (FFT). Metode ini dipilih karena kemampuannya untuk mengubah data deret waktu (time-domain) yang penuh dengan gangguan (noise) menjadi domain frekuensi, sehingga kita dapat mengisolasi dan menentukan periode osilasi utama matahari secara matematis dan objektif.</p>

### 2.1 Sunspot dan Siklus Matahari

<p>Bintik matahari atau <em>sunspots</em> adalah fenomena sementara yang muncul pada fotosfer Matahari sebagai bintik yang tampak lebih gelap dibandingkan area di sekitarnya. Secara fisik, sunspot merupakan manifestasi dari aktivitas magnetik Matahari yang sangat kuat. Kegelapan pada sunspot disebabkan oleh suhu permukaan yang lebih rendah (sekitar 3.000–4.500 K) dibandingkan dengan fotosfer sekitarnya (sekitar 5.780 K). Penurunan suhu ini terjadi karena konsentrasi fluks medan magnet yang sangat tinggi menghambat proses konveksi, yaitu mekanisme transportasi energi dari interior Matahari ke permukaannya. Akibatnya, energi panas tertahan di bawah permukaan dan area tersebut tampak lebih redup secara visual.</p>

<p>Sunspot biasanya muncul berpasangan dengan polaritas magnetik yang berlawanan. Kemunculan mereka tidaklah acak, melainkan mengikuti pola periodik yang sangat teratur. Fenomena ini pertama kali dipelajari secara mendalam oleh astronom amatir Jerman, Heinrich Schwabe, pada tahun 1843, yang kemudian dikenal sebagai Siklus Schwabe. Siklus ini memiliki periode rata-rata sekitar 11 tahun, meskipun secara historis durasinya dapat bervariasi antara 9 hingga 14 tahun. Selama satu siklus, jumlah bintik matahari akan meningkat dari titik minimum (<em>Solar Minimum</em>) menuju titik maksimum (<em>Solar Maximum</em>), sebelum akhirnya menurun kembali.</p>

<p>Aktivitas sunspot sangat penting dalam studi ruang angkasa karena berkorelasi langsung dengan emisi radiasi Matahari dan fenomena cuaca antariksa. Pada saat <em>Solar Maximum</em>, Matahari cenderung lebih aktif dalam menghasilkan <em>solar flares</em> dan <em>Coronal Mass Ejection</em> (CME). Ledakan energi ini dapat melepaskan partikel bermuatan tinggi ke seluruh sistem tata surya. Jika partikel ini berinteraksi dengan magnetosfer Bumi, dampaknya dapat mencakup gangguan pada sistem komunikasi satelit, fluktuasi pada jaringan listrik di permukaan Bumi, hingga munculnya fenomena aurora di wilayah kutub. Oleh karena itu, kemampuan untuk memprediksi dan menganalisis periode siklus ini melalui metode matematika seperti FFT sangat krusial dalam mitigasi risiko teknologi modern.</p>

### 2.2 Discrete Fourier Transform (DFT)

<p><em>Discrete Fourier Transform</em> (DFT) merupakan metode matematis yang digunakan untuk mengubah sinyal dari domain waktu menjadi domain frekuensi. Metode ini membantu dalam mengetahui frekuensi-frekuensi yang menyusun suatu sinyal sehingga pola periodik dapat dianalisis dengan lebih jelas. Dalam pengolahan sinyal digital, DFT sering digunakan untuk menganalisis data yang memiliki pola berulang. Penggunaan DFT sangat penting pada bidang fisika, telekomunikasi, dan astrofisika. Pada analisis data sunspot, DFT digunakan untuk mengidentifikasi pola periodik aktivitas matahari. Data jumlah sunspot yang awalnya berupa data terhadap waktu diubah menjadi representasi frekuensi sehingga frekuensi dominan dapat terlihat. Dengan metode ini, siklus aktivitas matahari dapat diketahui secara matematis dan lebih objektif. Analisis frekuensi juga membantu memisahkan sinyal utama dari <em>noise</em> pada data.</p>

<p>Perhitungan DFT secara langsung membutuhkan proses komputasi yang cukup besar apabila jumlah data sangat banyak. Oleh karena itu digunakan algoritma <em>Fast Fourier Transform</em> (FFT) yang merupakan metode perhitungan DFT yang lebih cepat dan efisien. FFT mampu mengurangi waktu komputasi secara signifikan tanpa mengurangi akurasi hasil transformasi. Metode ini sangat cocok digunakan pada dataset sunspot yang memiliki ribuan sampel data bulanan.</p>

<p>Secara matematis, DFT dinyatakan dengan persamaan berikut:</p>

$$X_k = \sum_{n=0}^{N-1} x_n \, e^{-i2\pi kn/N}$$

dengan:

- $X_k$ = hasil transformasi pada frekuensi ke-$(k)$
- $x_n$ = data sinyal pada domain waktu
- $N$ = jumlah sampel data
- $n$ = indeks waktu
- $k$ = indeks frekuensi

<p>Penerapan DFT dalam astrofisika sangat krusial karena fenomena alam seperti aktivitas magnetik matahari sering kali memiliki pola berulang yang tersembunyi di balik fluktuasi data yang tampak acak. Melalui transformasi ini, setiap variasi dalam jumlah bintik matahari diubah menjadi informasi mengenai seberapa kuat frekuensi tertentu hadir dalam keseluruhan data historis yang ada. Proses ini memungkinkan peneliti untuk memisahkan sinyal fisik yang nyata dari derau pengamatan, serta menentukan berapa kali fenomena tersebut berulang dalam satu satuan waktu tertentu. Agar hasil analisis valid, data harus diambil dengan interval pengambilan sampel yang tetap dan memenuhi syarat kritis yaitu frekuensi pengambilan sampel harus lebih tinggi dari dua kali frekuensi tertinggi yang ingin diamati, yang dikenal sebagai kriteria Nyquist-Shannon (Brigham, 1974). Dalam konteks data bintik matahari bulanan, interval satu bulan sudah sangat cukup untuk menangani frekuensi siklus matahari yang rendah, sehingga tidak terjadi kesalahan interpretasi frekuensi atau <em>aliasing</em> dalam hasil akhir.</p>

<p>Salah satu tahapan penting sebelum menjalankan DFT adalah penghapusan komponen rata-rata atau komponen arus searah (DC) agar spektrum frekuensi tidak didominasi oleh lonjakan nilai pada frekuensi nol yang tidak memiliki makna fisik terkait periodisitas. Jika langkah ini diabaikan, nilai rata-rata yang besar dari jumlah bintik matahari akan menutupi puncak-puncak kecil yang sebenarnya mengandung informasi tentang siklus aktivitas matahari itu sendiri. DFT bekerja dengan asumsi bahwa sinyal yang dianalisis bersifat periodik di luar rentang pengamatan, sehingga perlakuan terhadap nilai rata-rata dan ujung data sangat berpengaruh pada kualitas hasil spektrum yang dihasilkan (Press et al., 2007). Setelah transformasi dilakukan, hasilnya berupa nilai kompleks yang selanjutnya diolah untuk mendapatkan nilai magnitudo yang menunjukkan seberapa dominan suatu frekuensi dalam sinyal asli. Kemampuan DFT dalam mengkuantifikasi kontribusi setiap frekuensi inilah yang menjadi dasar utama dalam penentuan periode dominan siklus matahari sebesar 10,84 tahun yang diperoleh dalam studi kasus ini.</p>

### 2.3 Fast Fourier Transform (FFT)

<p><em>Fast Fourier Transform</em> (FFT) bukanlah sebuah transformasi baru yang berbeda dari <em>Discrete Fourier Transform</em> (DFT), melainkan sebuah algoritma efisien untuk menghitung DFT dengan kecepatan yang jauh lebih tinggi. Secara matematis, DFT mengubah sinyal dari domain waktu $x(n)$ ke domain frekuensi $X(k)$ menggunakan persamaan:</p>

$$X(k) = \sum_{n=0}^{N-1} x(n) \, e^{-j\frac{2\pi}{N}nk}$$

<p>Namun, perhitungan DFT secara langsung memerlukan kompleksitas komputasi sebesar $O(N^2)$. Artinya, jika kita memiliki 3.252 data seperti pada dataset sunspot ini, komputer harus melakukan sekitar 10,5 juta operasi perhitungan. Di sinilah FFT berperan; dengan menggunakan pendekatan <em>divide and conquer</em> (biasanya algoritma Cooley-Tukey), FFT mereduksi kompleksitas tersebut menjadi $O(N\log_2 N)$. Hal ini membuat proses komputasi ribuan kali lebih cepat dan efisien.</p>

<p>Dalam analisis siklus matahari, FFT bekerja dengan menguraikan data jumlah bintik matahari yang tampak acak menjadi penjumlahan sinyal-sinyal sinus dan cosinus dengan berbagai frekuensi. Setiap "puncak" yang muncul pada hasil spektrum FFT menunjukkan kekuatan (magnitudo) dari frekuensi tertentu yang menyusun sinyal tersebut. Frekuensi dengan puncak tertinggi (dominan) dianggap sebagai frekuensi fisik utama dari sistem yang diamati.</p>

<p>Salah satu aspek teknis yang krusial dalam FFT adalah pemahaman mengenai Frekuensi Nyquist, yang menyatakan bahwa frekuensi tertinggi yang dapat dideteksi secara akurat adalah setengah dari frekuensi pengambilan sampel ($f_s/2$). Dalam kasus ini, karena data diambil setiap bulan ($1/12$ tahun), maka frekuensi Nyquist adalah 6 siklus per tahun. Karena siklus matahari terjadi dalam rentang waktu tahunan yang panjang (frekuensi rendah), FFT sangat efektif untuk mengisolasi sinyal ini dari gangguan jangka pendek lainnya.</p>

### 2.4 Spektrum Daya

<p>Spektrum daya merupakan representasi yang menunjukkan kekuatan atau energi suatu sinyal terhadap frekuensi tertentu. Setelah data ditransformasikan menggunakan FFT, setiap frekuensi akan memiliki nilai amplitudo yang berbeda-beda. Nilai amplitudo tersebut menunjukkan seberapa besar kontribusi suatu frekuensi terhadap sinyal asli. Frekuensi dengan amplitudo tertinggi biasanya dianggap sebagai frekuensi dominan.</p>

<p>Dalam analisis aktivitas matahari, spektrum daya digunakan untuk menentukan siklus utama dari jumlah sunspot. Data sunspot yang memiliki <em>noise</em> dan fluktuasi acak akan lebih mudah dianalisis pada domain frekuensi dibandingkan domain waktu. Dengan melihat puncak pada grafik spektrum daya, pola periodik aktivitas matahari dapat diketahui dengan lebih jelas. Metode ini membantu memperoleh hasil analisis yang lebih objektif dan akurat.</p>

<p>Spektrum daya diperoleh dari magnitude hasil transformasi Fourier yang telah dinormalisasi terhadap jumlah data. Nilai magnitude tersebut menggambarkan energi masing-masing komponen frekuensi pada sinyal. Semakin tinggi nilai magnitude suatu frekuensi, maka semakin besar pengaruh frekuensi tersebut terhadap data. Pada penelitian ini, spektrum daya digunakan untuk menemukan frekuensi dominan siklus matahari.</p>

<p>Persamaan spektrum daya dinyatakan sebagai berikut:</p>

$$P(f) = \frac{2}{N} \left| Y_k \right|$$

Keterangan:

- $P(f)$ = magnitude atau kekuatan sinyal pada frekuensi tertentu
- $Y_k$ = hasil transformasi Fourier (FFT)
- $N$ = jumlah sampel data
- $|Y_k|$ = nilai absolut dari hasil FFT

### 2.5 Hubungan Frekuensi dan Periode

<p>Frekuensi dan periode merupakan dua konsep penting dalam analisis sinyal yang saling berhubungan satu sama lain. Frekuensi menunjukkan banyaknya pengulangan suatu siklus dalam satu satuan waktu, sedangkan periode menunjukkan waktu yang dibutuhkan untuk menyelesaikan satu siklus penuh. Dalam analisis data periodik seperti aktivitas matahari, hubungan antara frekuensi dan periode digunakan untuk menentukan lama siklus yang terjadi. Semakin besar frekuensi suatu sinyal, maka semakin kecil periode yang dimilikinya.</p>

<p>Pada analisis FFT, frekuensi dominan diperoleh dari puncak tertinggi pada spektrum daya. Frekuensi tersebut kemudian dikonversi menjadi periode agar hasil analisis lebih mudah dipahami secara fisik. Dalam penelitian ini, frekuensi dominan digunakan untuk menentukan lama siklus aktivitas matahari berdasarkan data sunspot historis. Periode yang diperoleh kemudian dibandingkan dengan teori Siklus Schwabe yang memiliki rata-rata sekitar 11 tahun.</p>

<p>Hubungan antara frekuensi dan periode dinyatakan dengan persamaan:</p>

$$T = \frac{1}{f}$$

dengan:

- $T$ = periode
- $f$ = frekuensi

<p>Persamaan tersebut menunjukkan bahwa periode berbanding terbalik dengan frekuensi. Jika nilai frekuensi semakin kecil, maka nilai periodenya akan semakin besar. Sebaliknya, jika frekuensi meningkat maka periode akan semakin pendek. Hubungan ini menjadi dasar dalam menentukan siklus dominan pada analisis data sunspot menggunakan FFT.</p>

### **3.3 Tahapan Penelitian**

<p>Penelitian ini dilakukan melalui beberapa tahapan sistematis yang dirancang untuk mengekstraksi sinyal periodik dari data <em>time-series</em> bintik matahari. Proses diawali dengan akuisisi dan pra-pemrosesan data melalui pemuatan dataset <code>sunspots.csv</code> menggunakan pustaka Pandas. Karena data waktu awal berupa format tanggal (YYYY-MM-DD), dilakukan konversi menjadi format <em>Fractional Year</em> agar dapat dipetakan secara matematis dalam sumbu X, yang kemudian diplot dalam domain waktu untuk mengamati fluktuasi jumlah bintik matahari secara visual. Selanjutnya, dilakukan penghilangan komponen DC (<em>DC Component Removal</em>) untuk mengatasi nilai rata-rata (<em>mean</em>) yang besar pada sinyal asli. Melalui proses <em>centering</em> menggunakan rumus:</p>

$$y_{\text{centered}} = y - \bar{y}$$

<p>lonjakan raksasa pada frekuensi 0 Hz dapat dihindari sehingga puncak spektrum yang muncul murni merepresentasikan osilasi fisis aktivitas matahari, bukan nilai rata-rata statistik.</p>

<p>Setelah data bersih, tahapan berlanjut ke transformasi domain frekuensi menggunakan algoritma <em>Fast Fourier Transform</em> (FFT) dengan parameter sampling &Delta;t = 1/12 tahun. Algoritma ini menghasilkan komponen frekuensi kompleks yang kemudian dikonversi menjadi <em>Normalized Power Spectrum</em> untuk memperoleh magnitudo nyata dari setiap bin frekuensi. Memasuki tahap analisis spektrum dan deteksi puncak, spektrum daya diplot pada rentang frekuensi 0 hingga 0,5 siklus/tahun untuk memfokuskan pengamatan pada area yang relevan. Frekuensi dominan (<em>f</em><sub>max</sub>) diidentifikasi menggunakan fungsi <code>np.argmax</code>, yang kemudian dikonversi menjadi satuan waktu fisik atau periode melalui hubungan matematis:</p>

$$T = \frac{1}{f_{\max}}$$

<p>Terakhir, dilakukan interpretasi dan penarikan kesimpulan dengan memvalidasi hasil periode yang ditemukan (10,84 tahun) terhadap literatur ilmiah mengenai Siklus Schwabe, guna memastikan bahwa hasil tersebut merupakan representasi variasi alami aktivitas matahari dan bukan merupakan artefak dari gangguan (<em>noise</em>) data.</p>

```python
# ================================================================= #
#  CASE STUDY 1: ANALISIS SIKLUS MATAHARI (VERSI FIX ERROR)          #
# ================================================================= #

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# --- 1. LOAD DATA ---
df = pd.read_csv('../data/sunspots.csv')

# Perbaikan: Mengubah kolom 'Date' menjadi format tanggal beneran, 
# lalu diubah menjadi angka tahun (misal: 1749.08) supaya bisa dihitung FFT
df['Date'] = pd.to_datetime(df['Date'])
t = df['Date'].dt.year + (df['Date'].dt.month - 1) / 12

# Mengambil kolom jumlah sunspot (kolom ke-3)
y = df['Monthly Mean Total Sunspot Number'].values

print(f"Berhasil memproses {len(t)} data dari tahun {t.min():.0f} sampai {t.max():.0f}!")

# --- 2. DC REMOVAL ---
# Sekarang 'y' sudah angka murni, jadi np.mean(y) tidak akan error lagi
y_centered = y - np.mean(y)

# --- 3. FFT & POWER SPECTRUM ---
N = len(y_centered)
dt = 1/12 # Jarak antar data adalah 1 bulan

Y_k = np.fft.fft(y_centered)
freq = np.fft.fftfreq(N, d=dt)

power = (2/N) * np.abs(Y_k)

# Ambil sisi positif saja
mask = freq > 0
f_pos = freq[mask]
p_pos = power[mask]

# --- 4. CARI PUNCAK ---
max_idx = np.argmax(p_pos)
f_max = f_pos[max_idx]
period = 1 / f_max

# --- 5. VISUALISASI ---
plt.figure(figsize=(12, 6))
plt.plot(f_pos, p_pos, color='blue', label='Spektrum Daya')
plt.scatter(f_max, p_pos[max_idx], color='red', s=100, label=f'Puncak: {period:.2f} Tahun')

plt.xlim(0, 0.5)
plt.xlabel('Frekuensi (Siklus per Tahun)')
plt.ylabel('Magnitude')
plt.title('Analisis Frekuensi Siklus Matahari')
plt.legend()
plt.grid(True)
plt.show()

print(f"Hasil Akhir: Periode Dominan adalah {period:.2f} Tahun")
```
